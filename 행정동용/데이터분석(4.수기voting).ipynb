## 패키지 참조

```python
from pandas import DataFrame,read_csv,read_excel,merge,melt,pivot_table,concat
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import sys
```

### 여가생활,그린지수 대상 최적의 군집 결과(양의지수) 교집합

```python
kmeans1=read_excel('./교집합결과/여가생활,그린지수voting1.xlsx')
dbscan1=read_excel('./교집합결과/여가생활,그린지수voting2.xlsx')
hery1=read_excel('./교집합결과/여가생활,그린지수voting3.xlsx')
```

#### concat 메서드 활용하여 데이터프레임끼리 묶어줌

```python
tmp=concat([kmeans1,dbscan1],ignore_index=True)
tmp.head()
```

```python
tmp1=concat([tmp,hery1],ignore_index=True)
tmp1.head()
```

#### 3개 군집 결과를 묶은 후 읍면동명으로 count하여 중복 선정(교집합)된 곳이 시범사업지 후보 대상

```python
tmp2=DataFrame(tmp1['읍면동명'].value_counts())
tmp2.to_excel('./교집합결과/여가생활,그린지수 보팅결과.xlsx')
tmp1['읍면동명'].value_counts().head(10)
```
